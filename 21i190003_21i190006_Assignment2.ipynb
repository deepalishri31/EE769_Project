```python
corr = df.corr()
corr.shape
```

```python
upper_tri_corr = np.triu(corr, k=1) 
#Since the correlation matrix is symmetric, hence changed it into an upper triangular matrix.
```

```python
upper_tri_corr
```

```python
from pandas.core.frame import DataFrame
train_updated = (df.drop(['Genotype','Treatment_Behavior'],axis = 1)).copy()
Dropping_cols = []
for i in range(corr.shape[0]):
  for j in range(corr.shape[1]):
    if abs(upper_tri_corr[i,j])>=0.89:
      #train_updated.drop([df.columns.values[i]],axis=1)
      Dropping_cols.append(df.columns.values[i])
    else:
      pass
print(Dropping_cols)
train_updated.drop(list(set(Dropping_cols)),axis = 1)
#convertying the data into dataframe and using the column as given in the training dataset.
train_updated = DataFrame(train_updated,columns = [col for col in df.columns.values[:-2] if col not in list(set(Dropping_cols))])
train_updated.shape 
```

We found the list of columns that we need to drop. We remove the variables which are correlated postively or negatively more than 0.89

And we found 6 such variables to remove.

```python
train_updated
```

```python
list(set(Dropping_cols))
```

The above given variables are redundant variables sinve these have significant positive or negative correlation with the other variables as clear from the correlation matrix.

###For checking the Balancing of classes:

```python
T=0 # Computing the number of mouse in the corresponding classes in Genotype.
C=0
for i in range(762):
  if df['Genotype'][i]=='Control':
    C+=1
  else:
    T+=1
print('Control: ',C)
print('Ts65Dn: ',T)
```

```python
#Plotting the barplot for visualizing the class balancing or imbalancing on types of 'Genotype'.
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 8)) 
# creating the bar plot
plt.bar(df['Genotype'].unique(), [C,T], color ='maroon')
plt.show()
```

```python
df['Treatment_Behavior'].unique() #Types of classes in Treatment behavior
```

```python
#Calculating the number of. mouse in each class on the basis of Treatment Behavior
M_CS = M_SC = S_CS = S_SC = 0
for i in range(762):
  if df['Treatment_Behavior'][i]== 'Memantine_C/S':
    M_CS+=1
  elif df['Treatment_Behavior'][i]== 'Memantine_S/C':
    M_SC +=1
  elif df['Treatment_Behavior'][i]== 'Saline_C/S':
    S_CS +=1
  else:
    S_SC +=1
print('Memantine_C/S : ',M_CS)
print('Memantine_S/C : ',M_SC)
print('Saline_C/S : ',S_CS)
print('Saline_S/C : ',S_SC)
```

```python
# creating the bar plot
fig = plt.figure(figsize = (10, 8)) 
plt.bar(df['Treatment_Behavior'].unique(), [M_CS,M_SC,S_CS,S_SC])
plt.show()
```

As clear from the above plots, here in both the cases, the clsses are almost balanced since None of the classes is dominating the others in singnificant way. Hence all the classes have a reasonable part in the distribution.

For Binary:

```python
from sklearn.linear_model import LogisticRegression #Imposrting model
params_ = {'l1_ratio' : np.arange(0.01,1,1e-2)} #Parameters for gridsearchcv method
LR = LogisticRegression(random_state=3, solver='saga', penalty='elasticnet',l1_ratio = 0.4)
model1_bin = GridSearchCV(LR,param_grid = {'l1_ratio' : np.arange(0.01,1,1e-2)},cv = 5,scoring = ['precision','balanced_accuracy','f1'],refit = 'balanced_accuracy',error_score='raise') #Used error_score as 'raise' to clearly state the error.
model1_bin.fit(df_train, y_binary) 
print('Best Hyperparameters for Binary Classification are: ',model1_bin.best_params_) # Best hyperparameters combination
print('Best Score is : ',model1_bin.best_score_) # For finding the best score from the hyperparameter tuning.
```

For MultiClass:

```python
from sklearn.preprocessing import LabelEncoder #Labvel Encoding the data for multiclass
le = LabelEncoder()
y_multiclass = le.fit_transform(y_multiclass) #LabelEncoding the multiclass data for use in the code.
```

```python
# define the multinomial logistic regression model with a default penalty
model1_MC = GridSearchCV(estimator = LogisticRegression(),param_grid ={'penalty':['elasticnet'],'l1_ratio' : np.arange(0.01,1,1e-2),'solver':['saga']},cv = 5,scoring = {'f1_micro','roc_auc_ovr'},refit = 'roc_auc_ovr',error_score='raise')
model1_MC.fit(df_train, y_multiclass)
print('Best Hyperparameters for multi Classification are: ',model1_MC.best_params_) #Best hyperparameters combination
print('Best Score is : ',model1_MC.best_score_) #Best Crossvalidated Score of the best estimator
```
